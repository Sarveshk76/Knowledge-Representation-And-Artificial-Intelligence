```python
import pandas as pd

df= pd.read_csv('social_network_ads.csv')
df.head()
```

```python
X = df.iloc[:,[2,3]].values
y = df.iloc[:,4].values 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.40,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(X_train)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
cls = KNeighborsClassifier(n_neighbors=8)
cls.fit(X_train,y_train)

y_pred = cls.predict(X_test)
y_test.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
res = confusion_matrix(y_test,y_pred)
print('Confusion Matrix:\n ',res)
res_1 = classification_report(y_test,y_pred)
print('Classification Report:\n ',res_1)
res_2 = accuracy_score(y_test,y_pred)
print('Accuracy Score:\n ',res_2)
```

```python
y_pred_train = cls.predict(X_train)

res_train = confusion_matrix(y_train,y_pred_train)
print('Confusion Matrix:\n ',res_train)
res_1_train = classification_report(y_train,y_pred_train)
print('Classification Report:\n ',res_1_train)
res_2_train = accuracy_score(y_train,y_pred_train)
print('Accuracy Score:\n ',res_2_train)
```
